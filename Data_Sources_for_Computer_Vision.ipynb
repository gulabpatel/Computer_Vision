#OpenML dataset

```
from sklearn.datasets import fetch_openml
faces = fetch_openml(name='UMIST_Faces_Cropped', version=1)
```

```
faces.data.shape  #There are 575 image and each with pixel 10304 pixels
```

```
faces.target
```

```
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(15,12))
for i in range(20):
  idx = np.random.randint(1,500)
  face = faces.data[idx]
  face_pixels = face.reshape(112, 92)
  plt.subplot(5, 4, i+1)
  plt.imshow(face_pixels)
  plt.axis('off')
```
